**Write a Pandas program to implement following operation**

*  to convert a NumPy array to a Pandas series

```python
import numpy as np
import pandas as pd

a=np.array([1,2,3])

b=pd.Series(a)

print(b)
```

*  to convert the first column of a DataFrame as a Series


```python
import pandas as pd

a={"name":['Mumtahin','hasan'],"spi":[9.14,9.81]}

df=pd.DataFrame(a)

print(df)
```

* to create the mean and standard deviation of the data of a given 
Series

```python
import pandas as pd

a=pd.Series([1,2,3,4,5,6,7,8,9])

mean=a.mean()

std=a.std()
print(mean)
print(std)
```

*  to sort a given Series

```python
import pandas as pd

a=pd.Series([8,9,5,3,1,2])

a=a.sort_values()

print(a)
```
