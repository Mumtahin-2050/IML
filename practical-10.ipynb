**Write a Pandas program to implement following operation**

* to find and drop the missing values from the given dataset

```python
import pandas as pd
import numpy as np
d={"Name":['Mumtahin',np.nan,'hasan']}

a=pd.DataFrame(d)

a=a.dropna()

print(a)
```

*  to remove the duplicates from the given dataset

```python
import pandas as pd
import numpy as np
d={"Name":['Mumtahin','Mumtahin','hasan']}

a=pd.DataFrame(d)

a=a.drop_duplicates()

print(a)
```
