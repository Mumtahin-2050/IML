```python
import pandas as pd
import matplotlib.pyplot as plt

file = pd.read_csv("/kaggle/input/nifty50-stock-market-data/BRITANNIA.csv")
start = "2000-01-03"
end = "2000-01-07"

file = file[(file["Date"] >= start) & (file["Date"] <= end)]

file = file.set_index("Date").sort_index(ascending = True)
file = file.reset_index()

data_open = file[["Date", "Open"]]
data_close = file[["Date", "Close"]]

plt.figure(figsize = (10,5))

plt.plot(data_open["Date"],data_open['Open'], color='green', marker='o', label='Open')
plt.plot(data_close["Date"],data_close['Close'], color='red', marker='o', label='Close')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Open and Close Prices')
plt.legend()
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
